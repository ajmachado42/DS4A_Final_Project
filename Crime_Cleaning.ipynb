# Filter FBI Hate Crime DF for top_thirty Cities

```python
#change column name
hc_df.rename(columns = {'race_ethnicity_ancestry':'hc_demo'}, inplace = True)
```

```python
# top_thirty city and state columns to respective lists
search_city = top_df["city"].tolist()
search_state = top_df["state"].tolist()
search = {"city": search_city, "state": search_state}
```

```python
search_city
```

```python
# filter for city-state match
# account for problem cities
top_hc_city = hc_df[hc_df["city"].isin(search_city)]
top_hc = top_hc_city[top_hc_city["state"].isin(search_state)]
top_hc = top_hc.drop(top_hc[(top_hc["city"] == "Jacksonville") & (top_hc["state"] == "Texas")].index)
top_hc = top_hc.drop(top_hc[(top_hc["city"] == "Louisville") & (top_hc["state"] == "Ohio")].index)
top_hc = top_hc.drop(top_hc[(top_hc["city"] == "Dallas") & (top_hc["state"] == "North Carolina")].index)
top_hc = top_hc.drop(top_hc[(top_hc["city"] == "Columbus") & (top_hc["state"] == "Indiana")].index)
top_hc = top_hc.drop(top_hc[(top_hc["city"] == "Washington") & (top_hc["state"] == "North Carolina")].index)
top_hc.reset_index()
```

```python
hc_top_df = pd.merge(top_df, top_hc[["city","hc_demo"]], on = "city", how = "left")
```

```python
def extraction(df, df_column):
    result_list = []
    
    if (top_df["city"] == df["city"]) and (top_df["state"] == df["state"]):
        c_to_list = df_column.tolist()
        result_list.append(c_to_list)
        
    return result_list
```
