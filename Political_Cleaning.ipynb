```python
def leg_party(x):
    """
    To find the political party counts for the legislators in each state
    Argument: state value from the leg_df["state"] dataset column
    Output: a count of legislators by political party
    """
    
    s = leg_df[leg_df["state"] == x]
    p = s.groupby(s["party"]).size()
    
    return p
```

# Governor and Mayor Cleaning

```python
# INCOMPLETE
def mayor_gov_party(city,state):
    """
    To find the political party of mayors in each city
    Argument: city value from the mayor_gov_df["city"] dataset column
    Output: the political party of the mayor for each city
    """
    c = mayor_gov_df[mayor_gov_df["city"] == "city"]
    mp = mayor_gov_df.loc[mayor_gov_df.city == "city", 'mayor_party']
    gp = mayor_gov_df.loc[mayor_gov_df.state == "state", 'gov_party']
    return mp, gp
```

```python
party_dict = {"D": "Democrat", "I":"Independent", "R":"Republican", "Democratic": "Democrat"}
mayor_gov_df["mayor_party"] = mayor_gov_df["mayor_party"].replace(party_dict)
mayor_gov_df["gov_party"] = mayor_gov_df["gov_party"].replace(party_dict)
mayor_gov_df
```
